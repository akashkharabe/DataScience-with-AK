```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

x = np.arange(0,10,0.1)
y = 2*x+5

plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,10,0.1)
y = 2*x+5

plt.plot(x,y)
plt.title("Line plot graph")
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.show()
```

```python
x = np.arange(0,10,1)
y = 2*x+5

fig=plt.figure(figsize=(15,5))
plt.plot(x,y, linewidth =2.0, linestyle=":",alpha = 0.5, color = "r", marker ="o")
plt.legend(["Line1"], loc ="best")
plt.grid(True)
plt.title("Line plot graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
x = np.arange(0,10,1)
y = 2*x+5
y1= 3*x+5

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("Graph1")

plt.subplot(1,2,2)
plt.plot(x,y1)
plt.title("Graph2")

plt.show()
```

#How to create Bar plot 

```python
data ={ "Apple": 20, "Mangoes":15, "Lemon":30, "Orange":10}
names = list(data.keys())
values= list(data.values())
fig =plt.figure(figsize=(10,5))
plt.bar(names, values)
plt.show()
```

```python
data ={ "Apple": 20, "Mangoes":15, "Lemon":30, "Orange":10}
names = list(data.keys())
values= list(data.values())
fig =plt.figure(figsize=(10,5))
plt.barh(names, values, color ="Red")
plt.title("Bar plot")
plt.xlabel("Fruits")
plt.ylabel("Quantity")
plt.show()
```

```python
a = [10,20,30,40,50,60,70,80,90]
b = [3,4,7,3,4,6,8,4,8]
x = [3,4,5,5,6,3,7,4,5]
fig=plt.figure(figsize=(10,5))
plt.scatter(a,b, c ="g",s = 300, edgecolors="y", alpha =0.7, marker ="o")
plt.scatter(a,x, c ="r",s = 350, edgecolors="b", alpha =0.5, marker ="4")
plt.legend(["b", "x"], loc='best')
plt.title("Scatter Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.savefig('ScatterPlot.png')
plt.grid(True)
plt.show()
```
